```python
import pandas as pd
import csv
import re


df = pd.DataFrame({'Swimming Pools': [], 'Waterbodies': [], 'Gardens':[], 'Boats':[], 'Parking Space':[], 'Solar Panels':[], 'Tennis Court':[] })
df.to_csv('HP.csv', index=True)

with open("detection_resultsV5.txt", "r") as file:
    lines = file.readlines()
swimmingpool_list=[]
gardens_list=[]
parkings_list=[]
solarpanels_list=[]
waterbodies_list=[]
tenniscourt_list=[]
boats_list=[]
v=0

for line in lines:
    if line.startswith("image"):
        parts = line.split(" ")
        v=v+1
        print("*****************v**********************",v)



        for i, j in enumerate(parts):
            #print("v",v)
            if j == "Swimming":
                print("Swimming Pools:", parts[i - 1])
                #swimmingpool_list.append(parts[i-1])
                df.at[v, 'Swimming Pools'] = parts[i-1]

                #print(swimmingpool_list)
            elif j == "gardens,":
                print("Gardens:", parts[i - 1])
                df.at[v, 'Gardens'] = parts[i-1]
            elif j == "parking,":
                print("Parking Spaces", parts[i - 1])
                df.at[v, 'Parking Space'] = parts[i-1]
            elif j == "solar":
                print("Solar Panels:", parts[i - 1])
                df.at[v, 'Solar Panels'] = parts[i-1]
            elif j == "waterbody,":
                print("Waterbodies:", parts[i - 1])
                df.at[v, 'Waterbodies'] = 1
            elif j == "tennis":
                print("Tennis Court:", parts[i - 1])
                df.at[v, 'Tennis Court'] = parts[i-1]
            elif j == "boats,":
                print("Boat Space:", parts[i - 1])
                df.at[v, 'Boats'] = parts[i-1]
                df.at[v, "Waterbodies"]=1
            #elif j=="22.6ms\n" or j==" 22.5ms\n":
              #print("***************One Cycle done********************")

```

```python
df = df.fillna(0)
df.to_csv('HP.csv', index=False)
df["Price"]=None
df
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
newdata = 20000000+ (df["Swimming Pools"] * 1000000) + (df["Tennis Court"] * 5000000) + (df["Parking Space"] * 1000000) + (df["Waterbodies"] * 30000000) + (df["Gardens"] * 500000) + (df["Boats"] * 2000000) + (df["Solar Panels"] * 500000)

df["Price"] = newdata

df["USD Price(M)"] = (df["Price"]/1000000)

df.to_csv(r"HP.csv", index=False)
```

```python
df['High end'] = (df['USD Price(M)'] < 30).astype(int)
df['Ultra High end'] = (df['USD Price(M)'] >= 30).astype(int)
```

```python
df.drop('USD Price', axis=1, inplace=True)
```

```python
df.to_csv("HP.csv",index=False)
```
